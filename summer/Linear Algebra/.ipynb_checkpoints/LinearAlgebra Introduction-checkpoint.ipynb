**Vectors**  - the building blocks of linear algebra

Vectors are useful in storing information made out of numerical data.  

For example:  
You (person A) have 3 apples and 4 oranges. This information can be stored in a vector - 3i + 4j.  
But in our data science pursuit, we will use different representation of vector ~  
$\begin{bmatrix} 3\\4 \end{bmatrix}$ *(Column vector)*

Along with the previous information, u also would like to track the information of Person B - who has 5 apples and 2 oranges.  
Well, that for that you need to merge the vectors ~ $\begin{bmatrix} 3&5\\4&2 \end{bmatrix}$, which is nothing but a *matrix*.  

Someone who have no knowledge of vectors, would have stored the same information in excel sheet, where each column represent a person and each row represent a fruit. Well, there's not at all any difference between the two representation. Please think through another example as well - Storing the properties of all Chemical Elements, like mass no., atomic no., valency, etc, How would you like to store this one?

Whole data science revolves around handling data, and therefore we need a structure to represent any chunk of data, store it in digital format and perform various mathematical operation in a most optimal way. That structure is **Vector**. 
 

```python
# representing the above vector using python
import numpy as np
A = np.array([[3],[4]])
print(A)
                                                                                 # we deal with matrix later..
```

**Vector operations**  

Vector addition and Scalar multiplication are the most talked about operations.

* **Vector Addition**:    
    Continuing with the previous example -  
    Possession of person A = $\begin{bmatrix} 3\\4 \end{bmatrix}$, and  
    Possession of person B = $\begin{bmatrix} 5\\2 \end{bmatrix}$.
    A and B give up their possessions to person C, who now will be having 8(3+5) apples and 6(4+2) oranges.  
    In a more vector perspective, the addition operation is shown like this :-  
    A + B = $\begin{bmatrix} 3\\4 \end{bmatrix}$ + $\begin{bmatrix} 5\\2 \end{bmatrix}$ = $\begin{bmatrix} 8\\6 \end{bmatrix}$
      

```python
# vector addition using python
A = np.array([[3],[4]])
B = np.array([[5],[2]])
print(A+B)
```

* **Scalar Multiplication**:  
    Similarly, if A's all possessins are doubled, he will have 6(2x3) apples and 9(2x4) oranges.  
    Representing it in a more vector notation ~  
    2 x $\begin{bmatrix} 3\\4 \end{bmatrix}$ = $\begin{bmatrix} 6\\8 \end{bmatrix}$

```python
# scalar multiplication using python
A = np.array([[3],[4]])
print(2*A)
```

**Geometric Representation of Vectors**  

To graphically represent an integer we need **Number Line**. <br>                                                                    Likewise, to represent a pair of integers we need **graph**.  <br> 
As you have guessed, we will represent the above vector **A = [ 3 4 ]** in a graph as well.
This is the same representation a physicist uses to represent a vector, saying every vector has 'a direction and a magnitude'.

```python
import matplotlib.pyplot as plt
```

```python
## graphical representation of vector 
plt.figure(figsize=(12,5))
plt.title('vector')
ax2 = plt.subplot(1,2,2)
ax2.set_ylabel('oranges')
ax2.set_xlabel('apples')
ax2.set_xlim([-1, 5])
ax2.set_ylim([-1, 5])
ax2.grid()
ax2.quiver(0, 0, 3, 0,color='r', angles='xy', scale_units='xy',scale=1)
ax2.quiver(3, 0, 0, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax2.plot(0,0,'o')

ax1 = plt.subplot(1,2,1)
ax1.set_title('vector')
ax1.set_ylabel('y-axis')
ax1.set_xlabel('x-axis')
ax1.set_xlim([-1, 5])
ax1.set_ylim([-1, 5])
ax1.grid()
ax1.quiver(0, 0, 3, 4,color='b', angles='xy', scale_units='xy',scale=1)
ax1.text(3.1,4, "A = [ 3, 4 ]", fontsize=15)
ax1.plot(0,0,'o')

plt.draw()
plt.show()
```

 * Vectors always begins from origin and ends at the coordinates
 * The information it carries :~
     * no. of apples
     * no. of oranges
 * Vector has Magnitude and Direction
 ---------------------------------------------------------------------

```python
# graphical representation of vector addition
plt.figure(figsize=(16,7))

ax1 = plt.subplot(1,2,1)
ax1.set_ylabel('Y-axis')
ax1.set_xlabel('X-axis')
ax1.set_title('Vector addition - [3,4] + [5,2]')
ax1.set_xlim([-1,9])
ax1.set_ylim([-1,9])
ax1.grid()
ax1.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax1.text(2,3.5, "A", fontsize=12)
ax1.quiver(0, 0, 5, 2,color='g', angles='xy', scale_units='xy',scale=1)
ax1.text(4,1, "B", fontsize=12)
ax1.quiver(0, 0, 8, 6,color='b', angles='xy', scale_units='xy',scale=1)
ax1.text(8,5.5, "A+B", fontsize=12)
ax1.plot(0,0,'o')


ax2 = plt.subplot(1,2,2)

ax2.set_title('using the law of Parallelogram for vector addition')
ax2.set_ylabel('Y-axis')
ax2.set_xlabel('X-axis')
ax2.set_xlim([-1,9])
ax2.set_ylim([-1,9])
ax2.grid()
ax2.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax2.text(2,3.5, "A", fontsize=12)
ax2.quiver(0, 0, 5, 2,color='g', angles='xy', scale_units='xy',scale=1)
ax2.text(4,1, "B", fontsize=12)
ax2.quiver(5, 2, 3, 4,color='y', angles='xy', scale_units='xy',scale=1, hatch='ooo')
ax2.text(6,3, "+ A", fontsize=12)
ax2.quiver(0, 0, 8, 6,color='b', angles='xy', scale_units='xy',scale=1)
ax2.text(6.5,5.5, "A+B", fontsize=12)
ax2.plot(0,0,'o')
plt.draw()
plt.show()
```

    Vector addition 'may' change the direction of a vector.                                                                        
--------------------------------------------

```python
# graphical representation of scalar multiplication
plt.figure(figsize=(16,7))

ax1 = plt.subplot(1,2,1)
ax1.set_ylabel('Y-axis')
ax1.set_xlabel('X-axis')
ax1.set_title('Scalar Multiplication - 2 x [3,4]')
ax1.set_xlim([-1,9])
ax1.set_ylim([-1,9])
ax1.grid()
ax1.quiver(0.1, 0, 6, 8,color='brown', angles='xy', scale_units='xy',scale=1)
ax1.text(6.2,8, "2xA", fontsize=12)

ax1.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax1.text(2,3.5, "A", fontsize=12)


ax2 = plt.subplot(1,2,2)
ax2.set_title('2 x A = A + A')
ax2.set_ylabel('Y-axis')
ax2.set_xlabel('X-axis')
ax2.set_xlim([-1,9])
ax2.set_ylim([-1,9])
ax2.grid()
ax2.quiver(0.1, 0, 6, 8,color='brown', angles='xy', scale_units='xy',scale=1)
ax2.text(6.2,8, "2xA", fontsize=12)

ax2.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax2.text(2,3.5, "A", fontsize=12)
ax2.quiver(3, 4, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax2.text(3.8,6, "+ A", fontsize=12)
ax2.plot(3,4,'o',color='black',)
plt.draw()
plt.show()
```

                 Scalar multiplication does not change the line.
-------------------------------------------------------------

* **Vector Subtraction**
    * Vector subtraction between 2 vectors can be thought as combination of vector addition and scalar multiplication ->  A - B = A + ( -1 x B )  
      therefore,    
      A - B = $\begin{bmatrix} 3\\4 \end{bmatrix}$ - $\begin{bmatrix} 5\\2 \end{bmatrix}$ = $\begin{bmatrix} -2\\2 \end{bmatrix}$

```python
# graphical representation of vector subtraction
plt.figure(figsize=(15,5))

ax1 = plt.subplot(1,2,1)
ax1.set_ylabel('Y-axis')
ax1.set_xlabel('X-axis')
ax1.set_title('Vector subtraction -> [3,4] - [5,2]')
ax1.set_xlim([-3,6])
ax1.set_ylim([-1,5])
ax1.grid()
ax1.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax1.text(2,3.5, "A", fontsize=12)
ax1.quiver(0, 0, 5, 2,color='g', angles='xy', scale_units='xy',scale=1)
ax1.text(4,1, "B", fontsize=12)
ax1.quiver(0, 0, -2, 2,color='b', angles='xy', scale_units='xy',scale=1)
ax1.text(-2,1, "A - B", fontsize=12)
ax1.plot(0,0,'o')


ax2 = plt.subplot(1,2,2)
ax2.set_title('B + (A - B) = A  (law of parallelogram)')
ax2.set_ylabel('Y-axis')
ax2.set_xlabel('X-axis')
ax2.set_xlim([-3,6])
ax2.set_ylim([-1,5])
ax2.grid()
ax2.quiver(0, 0, 3, 4,color='y', angles='xy', scale_units='xy',scale=1)
ax2.text(2,3.5, "A", fontsize=12)

ax2.quiver(0, 0, 5, 2,color='g', angles='xy', scale_units='xy',scale=1)
ax2.text(4,1, "B", fontsize=12)

ax2.quiver(5, 2, -2, 2,color='gray', angles='xy', scale_units='xy',scale=1, hatch='ooo')
ax2.text(4.1, 3.1, "+ (A - B)", fontsize=13)

ax2.quiver(0, 0, -2, 2,color='b', angles='xy', scale_units='xy',scale=1)
ax2.text(-2,1, "A - B", fontsize=13)
ax2.plot(0,0,'o')
plt.draw()
plt.show()
```

          Vector subtraction also may change the direction of the vector

### Linear Combination

Linear Combination is combination of Vector Addition and Scalar Multiplication.  
If $v1,...,vn$ are vectors and $a1,...,an$ are scalars, then the **Linear combination** of those vectors with those scalars as coefficients is
$a_1v_1+a_2v2+a_3v _3+\cdots +a_nv _n.$

In $R^2$ space:  
* LC of 1 vector forms a line containing that vector.
* LC of 2 linearly independent vectors forms the $R^2$ space.

In $R^3$ space:  
* LC of 1 vector forms a line containing that vector.
* LC of 2 linearly independent vectors forms a plane containing those 2 vectors.
* LC of 3 linearly independent vectors forms the $R^3$ space.

In $R^4$ space:  
* LC of 1 vector forms a line containing that vector.
* LC of 2 linearly independent vectors forms the plane contaning those 2 vectors.
* LC of 3 linearly independent vectors forms 3D Hyperpalne.
* LC of 4 linearly independent vectors forms the $R^4$ space.
