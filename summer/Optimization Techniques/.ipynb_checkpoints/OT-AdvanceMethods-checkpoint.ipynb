## Optimization Techniques

    Optimization techniques are useful in finding the optimum solution or unconstrained maxima or minima of functions.
    In Machine Learning, we are always in need to minimize the error function in a most optimal way.

## Direct Search Method

* Direct search is a method for solving optimization problems that does not require any information about the **gradient** of the objective function.
* Unlike more traditional optimization methods that use information about the gradient or higher derivatives to search for an optimal point, a direct search algorithm searches a set of points around the current point, looking for one where the value of the objective function is lower than the value at the current point.
* Direct search method can be used to solve problems for which the objective function is **not differentiable**, or is **not even continuous**.

Let's learn the algorithm with an example~  

Consider the function:  
f(x) =$\begin{cases}3x & ,0<=x<=2\\ \frac{20-x}{3} &,2<=x<=10\end{cases}$

```python
# example function
def fx(x):
    if(0<=x and x<=2):
        return 3*x
    elif(2<=x and x<=10):
        return (20-x)/3
    else:
        return None

func = np.vectorize(fx)

x=np.linspace(0,10)
y=func(x)
plt.title('f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y)
plt.show()
```

**Algorithm for finding Maximum**:  
* We start with 2 random points on x-axis- x_l and x_r,from the domain of f(x), and satisfies the condition x_l < x_r.  
* Iterate through the following steps until x_l == x_r:  
    * Find out 2 points x1, x2, such that x_l <= x1 <= x2 <= x_r  
    * if f(x1) > f(x2), then change x_r = x2
    * else if f(x1) < f(x2), then change x_l = x1
    * else if f(x1) == f(x2), then change both x_l = x1 and x_r = x2
* At the end of iteration when x_l == x_r, **x_l / x_r** is the point for which f(x) is maximum, between the first 2 ramdom points taken before the start of iteration.

The below example shows the situation of one such iteration ~  
Where, x_l = 0.2 and x_r = 9.6.  
We need to find 2 points x1, x2 such that x_l <= x1 <= x2 <= x_r.  
Let, x1 = 1.5 and x2 = 8.  
f(x1) = f(1.5) = 4.5  
f(x2) = f(8) = 4  
As we can see f(x1) > f(x2), we will change x_r = x2 = 8  
Therfore, at the end of the this iteration x_l, x_r = 0.2, 8

```python
plt.title('At ith iteration')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot((0.4,0.4),(0,func(0.4)), color='k',linestyle='--' )
plt.plot((9.6,9.6),(0,func(9.6)), color='k',linestyle='--')

plt.plot((1.5,1.5),(0,func(1.5)), color='gray',linestyle='-.')
plt.plot((8,8),(0,func(8)), color='gray',linestyle='-.')

plt.plot(x,y, label='f(x)')
plt.text(0.5,0.5, 'x_l')
plt.text(9,2, 'x_r')

plt.text(1.6,2.5, 'x1')
plt.text(7.5,2.1, 'x2')

plt.axhline(y = func(1.5), color = 'g', linestyle = ':')
plt.axhline(y = func(8), color = 'g', linestyle = ':')

plt.text(1.6,4.6, 'f(x1)=4.5')
plt.text(8,4.1, 'f(x2)=4')

plt.legend()
plt.show()
```

```python
plt.title('After ith iteration')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot((0.4,0.4),(0,func(0.4)), color='k',linestyle='--' )
plt.plot((8,8),(0,func(8)), color='k',linestyle='--')
plt.text(0.5,0.5, 'x_l')
plt.text(7.4,2.1, 'x_r')
plt.plot(x,y, label='f(x)')
plt.legend()
plt.show()
```

Before we find the maxima of f(x), we need a definite method to find x1 and x2 at any iteration. It is evident from the above example, that if we select x1, x2 too far from x_l, x_r, we may miss the maxima.  
There are 2 methods for find x1, x2.  
* **Dichotomous Method**:
    - $x_1 = \frac{1}{2} (x_r + x_l -\Delta)$  
    - $x_2 = \frac{1}{2} (x_r + x_l +\Delta)$ 

  
* **Golden Section Method**:
    - $x_1 = x_r - \frac{\sqrt{5}-1}{2} (x_r - x_l)$  
    - $x_2 = x_l + \frac{\sqrt{5}-1}{2} (x_r - x_l)$

```python
# dicotomous method
def dico(x_l,x_r,delta):
    x1 =(x_l+x_r-delta)/2
    x2 =(x_l+x_r+delta)/2
    return (x1,x2)

#golden section method
def gsm(x_l,x_r,delta):
    x1 =x_r - ((math.sqrt(5)-1)*(x_r-x_l)/2)
    x2 =x_l + ((math.sqrt(5)-1)*(x_r-x_l)/2)
    return (x1,x2)

#maximizer
def maxim(func=None,x_l=0,x_r=0,method='gsm'):
    if(func==None):
        return 'x'
    met=gsm
    if(method=='dico'):
        met=dico
    i=0
    
    plt.title(f'Finding maxima of f(x) using {method}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    f = np.vectorize(func)
    x=np.linspace(x_l,x_r)
    y=f(x)
    plt.plot(x,y)
    
    while(x_l<x_r):
        
        plt.plot((x_l,x_l),(0,func(x_l)), color='gray',linestyle='--' )
        plt.plot((x_r,x_r),(0,func(x_r)), color='gray',linestyle='--')
        i+=1
        x1,x2 = met(x_l,x_r,0.1)
        if(func(x1)>func(x2)):
            if(x_r==x2):
                break
            x_r=x2
        elif(func(x1)<func(x2)):
            if(x_l==x1):
                break
            x_l=x1
        else:
            if(x_l==x1):
                break
            x_l=x1
            x_r=x2
    plt.plot((x_l,x_l),(0,func(x_l)), color='r' )
    plt.show()
    print('Total Iterations-',i)
    print(f'f(x) is maximum at x={x_l:.5f}')
```

```python
maxim(func=fx,x_l=0, x_r=10,method='gsm')
```

```python
maxim(func=fx,x_l=0, x_r=10,method='dico')
```

**Algorithm for finding Minimum**:  
* It is pretty similar, with minute changes in the condition.
* We start with 2 random points on x-axis- x_l and x_r,from the domain of f(x), and satisfies the condition x_l < x_r.  
* Iterate through the following steps until x_l == x_r:  
    * Find out 2 points x1, x2, such that x_l <= x1 <= x2 <= x_r  
    * if f(x1) > f(x2), then change x_l = x1
    * else if f(x1) < f(x2), then change x_r = x2
    * else if f(x1) == f(x2), then change both x_l = x1 and x_r = x2
* At the end of iteration when x_l == x_r, **x_l / x_r** is the point for which f(x) is manimum, between the first 2 random points taken before the start of iteration.
* Note that, for finding x1 and x2, any of the methods(gsm/dico) can be used.

## Gradient Search Method

Gradient Search method is another iterative optimization algorithm for finding a local maximum/minimum of a **differentiable function**. 
A gradient measures how much the output of a function changes if you change the inputs a little bit.   
**Types~**
* Gradient Ascent - for finding Maximum  
* Gradient Descent - for finding Minimum  

*As you know that this popular gradient descent method is used in minimizing error function in various machine learning algorithms.*

Lets define a differentiable function:   *(the previous function is not differentiable at x=2)*  
$f(x) = \frac{sin(x)}{x}$  
and, 
$f^{'}(x) = \frac{xCos(x) - Sin(x)}{x^2}$  

```python
# example function
fx = lambda x: math.sin(x)/x if x!=0 else 1
dfx = lambda x: (x*math.cos(x)-math.sin(x))/(x**2)
func = np.vectorize(fx)

x=np.linspace(-8,8)
y=func(x)
plt.title('f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,y)
plt.show()
```

Algorithm for **Gradient Descent**:  
1. We start with an random value on x-axis $x_k$ 
2. We generate a point $x_{k+1} = x_k - (f^{'}(x_k) \cdot \delta)$ [$\delta$-learning rate]
3. If $x_{k+1} \neq x_k$, make $x_k= x_{k+1}$ and repeat from step 2. 
4. The loop stops when $x_{k+1}$ is almost equal to $x_k$, which is nothing but the point for which f(x) is minimum. 

* Lets, start from a random point on $x_k$=1.  
* Consider, $\delta=0.8$
* At first iteration,  
    * $f^{'}(x_k)=f^{'}(1)=-0.30116$
    * $x_{k+1} =x_k - (f^{'}(x_k) \cdot \delta) = 1 - (-0.30116\times 0.8)$  
    * $x_{k+1} = 1.24093$


```python
1 - (dfx(1)*0.8)
```

```python
x=np.linspace(-8,8)
y=func(x)
xk1= 1 - (dfx(1)*0.8)
plt.title('At first iteration..')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot((1,1),(0,func(1)), color='k',linestyle='--' )
plt.plot((xk1,xk1),(0,func(xk1)), color='gray',linestyle=':' )
plt.text(0,0.5, 'x_k')
plt.text(1.4,0.3, 'x_k+1')
plt.plot(x,y, color='g')
plt.show()
```

```python
def gradDescent(fx, dfx, xk, delta=0.8):
    xk1 = xk - (dfx(xk)*delta)
    x=np.linspace(-8,8)
    y=np.vectorize(fx)(x)
    plt.title('Finding the minima(local)..')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(x,y, color='g')
    i=0
    while(abs(xk1-xk)>0.01):
        plt.plot(xk,func(xk), color='b', marker='.')
        xk=xk1
        xk1 = xk - (dfx(xk)*delta)
        i+=1
    plt.plot(xk1,func(xk1), color='r', marker='s' )
    plt.show()
    print(f'Total no. of iterations - {i}')
    print(f'Point of minima - {(xk+xk1)/2:.3f}')
```

Now, lets find out the minima, starting from x=1

```python
gradDescent(fx, dfx, 1)
```

Lets what happens when we start from x= -1

```python
gradDescent(fx, dfx, -1)
```

*Therfore, one can imagine Gradient descent method as how a rain drop finds its way, from the place where it has dropped, to a local pond.*

Algorithm for **Gradient Ascent**:  
1. We start with an random value on x-axis $x_k$ 
2. We generate a point $x_{k+1} = x_k + (f^{'}(x_k) \cdot \delta)$ [$\delta$-learning rate]
3. If $x_{k+1} \neq x_k$, make $x_k= x_{k+1}$ and repeat from step 2. 
4. The loop stops when $x_{k+1}$ is almost equal to $x_k$, which is nothing but the point for which f(x) is maximum. 

Remember - 
* Gradient Method works only on diffrentiable function.
* Gradient Method searches for local minima/maxima, unlike direct search method which searches for global minima/maxima.

#### For 2 dimensional function ~  

Gradient descent formula:  
$(x_{k+1},y_{k+2}) = (x_k,y_k) - \delta \cdot f^{'}(x_k,y_k) $
  
 Gradient ascent formula:  
$(x_{k+1},y_{k+2}) = (x_k,y_k) + \delta \cdot f^{'}(x_k,y_k) $

Let's work with 2-dimensional function:  
$f(x,y) = Sin(\pi x)Sin(\pi y)$  

```python
fx = lambda x,y: math.sin(math.pi*x)*math.sin(math.pi*y)
dfx = lambda x, y: np.array((math.pi*math.cos(math.pi*x)*math.sin(math.pi*y),math.pi*math.sin(math.pi*x)*math.cos(math.pi*y)))
```

```python
from mpl_toolkits import mplot3d

plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
x = np.linspace(-1, 1)
y = np.linspace(-1, 1)
X, Y = np.meshgrid(x, y)
Z = np.vectorize(fx)(X,Y)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='autumn_r', edgecolor=None)
plt.show()
```

```python
def gradAscent3d(fx, dfx, xk,yk, delta=0.1):
    
    xk1, yk1 = np.array((xk,yk)) + dfx(xk,yk)*delta
    xk1, yk1 =round(xk1,6), round(yk1,6)
    plt.figure(figsize=(6,6))
    ax = plt.axes(projection='3d')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    x = np.linspace(-1, 1)
    y = np.linspace(-1, 1)
    X, Y = np.meshgrid(x, y)
    Z = np.vectorize(fx)(X,Y)
    
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='autumn_r', edgecolor='none')
    
    i=0
    while(abs(xk1-xk)>0.01 or abs(yk-yk1)>0.01):
        ax.plot(xk,yk,fx(xk,yk)+0.1, color='b', marker='.' )
        xk,yk = xk1, yk1
        xk1, yk1 = np.array((xk,yk)) + dfx(xk,yk)*delta
        xk1, yk1 =round(xk1,6), round(yk1,6)
        i+=1
    plt.show()
    print(f'Total no. of iterations - {i}')
    print(f'Point of maxima - ({xk1:.3f},{yk1:.3f})')
```

```python
gradAscent3d(fx, dfx, 0.5,0)
```
