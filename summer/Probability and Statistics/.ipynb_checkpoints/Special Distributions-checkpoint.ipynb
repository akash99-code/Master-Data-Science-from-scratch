# Special Distributions
-----------------

Probability distribution is the mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment. This Probability mass/density function comes with various characteristics, depending on the type of experiment it summarizes.

### Types of Special Distribution
* Discrete
    * Bernouli
    * Binomial
    * Poisson
    * Uniform
          
          
* Continuous
    * Exponential
    * Uniform
    * Normal
    

1. **Bernouli Distribution**:<br/> 
    Bernoulli distribution is the discrete probability distribution of a Random Variable, which takes only two possibilities - success(**1**) and failure(**0**).  
    Example - When a coin is tossed and the outcome is head(*success*), random variable $X$ takes 1 else 0.  
    Let $X$ be a discrete random variable having Bernouli Distrbution, whose probability of success is **p**,  
    then its **pmf** is  
    $f(x)=\begin{cases}p & ,X=1\\(1-p) & ,X=0\end{cases}$  
      
        
  * Expectation of Bernouli    
    $E[X] = p\hspace{2mm}[1\times p + 0\times(1-p)]$ 
      
      
  * Second Moment of Bernouli  
    $E[X^2] = p\hspace{2mm}[1^2\times p + 0^2\times(1-p)]$
        
        
  * Variance of Bernouli  
    $Var[X] = (1-p)^2\times p \hspace{2mm}+\hspace{2mm} (0-p)^2 \times (1-p)$  
    $\hspace{1.3cm}=(1-p)p\{1-p+p\}$  
    $\hspace{1.3cm}=pq\hspace{2mm}$[q=(1-p)]  
    Also,  
    $Var[X] = E[X^2] - E[X]^2$  
    $\hspace{1.3cm}=p-p^2$  
    $\hspace{1.3cm}=pq$ 
    
  * Moment Generating Function of Bernouli  
    $\varphi_X (t) = E[e^{tX}]$  
    $\hspace{1.3cm}= e^{t\cdot 1}p\hspace{2mm}+\hspace{2mm}e^{t\cdot0}(1-p)$  
    $\hspace{1.3cm}= pe^t+q\hspace{2mm}$[q=(1-p)]  

2. **Binomial Distribution: X~B(n,p)**  
Binomial Distribution is the discrete probability distribution of Random Variable X, which takes the value as total number of successes achieved when a boolean experiment(*experiment which has only two outcomes*) when executed for n times.  
Example - When a coin tossed for 10 times out of which we get 6 heads(*success*), random variable $X$ for the entire experiment takes the value $X=6$. We can also say **Binomial** R.V. is prepared from repeated **Bernouli** R.V. 
  
    Let $X$ be a Binomial R.V, whose probability of success of a single trial is $p$ and experimented for $n$ times,**X~B(n,p)**,   then it's **pmf** can be defined as:  
$f(x) = ~^nC_x p^xq^{n-x} ,~~for~ x=[0,1,2,...,n]\hspace{2mm}$[$x$=no. of success, failure $q=(1-p)$]  
  
* Expectation of Binomial    
    $E[X] = \sum^n_{x=0} x \cdot ~ {n\choose x} p^xq^{n-x}$  
    $\hspace{1.3cm}= \sum^n_{x=1} x \cdot \frac{n}{x} ~{n-1\choose x-1} p^xq^{n-x}$  
    $\hspace{1.3cm}= np\cdot \sum^n_{x=1} ~{n-1\choose x-1} p^{x-1}q^{(n-1)-(x-1)}$  
    $\hspace{1.3cm}= np\cdot (p+q)^{n-1}\hspace{4mm}$[p+q=1]  
    $\hspace{1.3cm}= np$  
    We can also derive like,   
    $E[X] = n \cdot E[X_{bernouli}]$  
    $\hspace{1.3cm}= np$  
      
      
* Second Moment of Binomial  
    $E[X^2] = \sum^n_{x=0} x^2 \cdot ~ {n\choose x} p^xq^{n-x}$  
    $\hspace{1.3cm}= \sum^n_{x=1} (x^2-x+x) \cdot \frac{n}{x} \hspace{2mm}{n-1\choose x-1} p^xq^{n-x}$  
    $\hspace{1.3cm}= \sum^n_{x=1} x^2-x \cdot \frac{n}{x} \hspace{2mm}{n-1\choose x-1} p^xq^{n-x} + \sum^n_{x=1} x \cdot \frac{n}{x} ~{n-1\choose x-1} p^xq^{n-x}$  
    $\hspace{1.3cm}= \sum^n_{x=1} x(x-1) \cdot \frac{n(n-1)}{x(x-1)} ~{n-2\choose x-2} p^xq^{n-x} + \sum^n_{x=1} np ~{n-1\choose x-1} p^{(x-1)}q^{(n-1)-(x-1)}$  
    $\hspace{1.3cm}= n(n-1)p^2 + np$    
    $\hspace{1.3cm}= n^2p^2\hspace{2mm}-\hspace{2mm}np^2~+~np$  
    $\hspace{1.3cm}= n^2p^2\hspace{2mm}+\hspace{2mm}np(1-p)$  
        
        
* Variance of Binomial  
    $Var[X] = E[X^2] - E[X]^2$  
    $\hspace{1.3cm}=n^2p^2\hspace{2mm}+\hspace{2mm}np(1-p)\hspace{2mm}-\hspace{2mm}(np)^2$  
    $\hspace{1.3cm}=npq\hspace{0.2cm}$[1-p=q]  
    Can be also derived like,  
    $Var[X] = n \cdot Var[X_{bernouli}]$  
    $\hspace{1.3cm}=npq$  
     
    
* Moment Generating Function of Binomial  
    $\varphi_X (t) = E[e^{tX}]$  
    $\hspace{1.3cm}= \sum_x e^{tx}\cdot \hspace{2mm}^nC_x p^xq^{(n-x)}$  
    $\hspace{1.3cm}= \sum_x \hspace{2mm}^nC_x (e^tp)^xq^{(n-x)}$  
    $\hspace{1.3cm}= (pe^t+q)^n$

3. **Poisson Distribution:X~P($\lambda$)**  
Poisson Distribution is the discrete probability distribution of Random Variable X, whose value represents the total number of times a event that has occurred in a given interval of time/space. These events do not effect of each other and their frequency of occurence has a fixed **mean->**($\lambda$).  
Example - A call center receives an average of 200 calls in a day*(interval)*. Therefore, our mean $\lambda=200$. Suppose, today the call center received 210 calls, then our R.V $X=210$. 
  
    Let $X$ be a Poisson R.V, whose mean is $\lambda$,**X~P($\lambda$)**,   then it's **pmf** can be defined as:  
$f(x) = \frac{e^{-\lambda}\lambda^x}{x!} ~~for~x=[0,1,2,....]$  
  
* Expectation of Poisson    
    $E[X] = \sum^\infty_{x=0} x \cdot ~ \frac{e^{-\lambda}\lambda^x}{x!}$  
    $\hspace{1.3cm}= e^{-\lambda}\sum^\infty_{x=1} \frac{\lambda \cdot \lambda^(x-1)}{(x-1)!}$  
    $\hspace{1.3cm}= \lambda e^{-\lambda}\sum^\infty_{x=1} \frac{\lambda^(x-1)}{(x-1)!}\hspace{0.4cm}$[$e^x$ series]  
    $\hspace{1.3cm}=\lambda e^{-\lambda}\cdot e^{-\lambda}$  
    $\hspace{1.3cm}= \lambda$ =**Mean**(which was already declared)  
      
      
* Second Moment of Poisson    
    $E[X^2] = \sum^\infty_{x=0} x^2 \cdot ~ \frac{e^{-\lambda}\lambda^x}{x!}$  
    $\hspace{1.3cm}= e^{-\lambda}\{\sum^\infty_{x=1} (x^2-x+x)\cdot \frac{\lambda^x}{x!}\}$  
    $\hspace{1.3cm}= e^{-\lambda}\{\sum^\infty_{x=1} x(x-1)\cdot\frac{\lambda^x}{x!}\hspace{0.3cm} + \hspace{0.3cm}\sum^\infty_{x=1} x\cdot\frac{\lambda^x}{x!}\}$    
    $\hspace{1.3cm}= e^{-\lambda}\{\lambda^2\sum^\infty_{x=2} \frac{\lambda^{(x-2)}}{(x-2)!}\hspace{0.3cm}+\hspace{0.3cm}\lambda\sum^\infty_{x=1} \frac{\lambda^{(x-1)}}{(x-1)!}\}$  
    $\hspace{1.3cm}=e^{-\lambda}\{\lambda^2e^{-\lambda}\hspace{0.3cm} +\hspace{0.3cm}\lambda e^{-\lambda}\}$  
    $\hspace{1.3cm}= \lambda^2 +\lambda$  
      
      
* Variance of Poisson    
    $Var[X] = E[X^2] - E[X]^2$  
    $\hspace{1.3cm}= (\lambda^2 +\lambda) - \lambda^2$  
    $\hspace{1.3cm}= \lambda$  
      
* Moment Generating function of Poisson    
    $\varphi_X (t) = E[e^{tX}]$  
    $~\hspace{1.3cm}= \sum_x e^{tx}\frac{e^{-\lambda}\lambda^x}{x!}$  
    $~\hspace{1.3cm}= e^{-\lambda} \cdot \sum_x \frac{(e^t\lambda)^x}{x!}$  
    $~\hspace{1.3cm}= e^{-\lambda} \cdot e^{\lambda e^t}$  
    $~\hspace{1.3cm}= e^{\lambda(e^t-1)}$

4. **Exponential Distribution:X~E($\lambda$)**  
Exponential Distribution is the continuous probability distribution of the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate($\frac{1}{\lambda}$). Therfore, Random Variable of Exponential distribution is the continuous version of that of Poisson distribution, and also with a fixed mean $1/\lambda$.   
Example - Let R.V. $X$ takes the value as hours for which a particular CFL bulb will last. The company claims the bulb to have average lifespan of 8000 hours. Thus, $1/\lambda$ = 8000 or $\lambda$ = 0.000125.  

    Let $X$ be a continuous Uniform R.V, whose constant mean rate = $1/\lambda$, then it's **pdf** can be defined as:  
$f(x) = \begin{cases}\lambda e^{-\lambda x} & x>=0 \\ 0 & x<0\end{cases}$   
  
* Expectation of exponential    
    $E[X] = \int_{x=0}^\infty x \cdot ~ \lambda e^{-\lambda x} dx$  
    $\hspace{1.3cm}= x\int \lambda e^{-\lambda x}dx ~~~-~~~ \int \frac{d}{dx}x \int \lambda e^{-\lambda x}dx~dx$  
    $\hspace{1.3cm}= -xe^{-\lambda x} ~~~+~~~ \int e^{-\lambda x}$  
    $\hspace{1.3cm}=- \left[ \frac{e^{-\lambda x}}{\lambda}\right]_0^\infty$      
    $\hspace{1.3cm}= \frac{1}{\lambda}$
      
      
* Second moment of exponential  
    $E[X^2] = \int_{x=0}^\infty x^2 \cdot ~ \lambda e^{-\lambda x} dx$  
    $\hspace{1.3cm}= \frac{2}{\lambda^2}$  
    
      
* Variance of exponential  
    $Var[X] = E[X^2] - E[X]^2$  
    $\hspace{1.3cm}= \frac{1}{\lambda^2}$  
      
* Moment Generating function of exponential  
    $\varphi_X (t) = \int_0^\infty e^{tx} \lambda e^{-\lambda x} dx$  
    $\hspace{1.3cm}= \frac{\lambda}{\lambda-t}$, for $\lambda$>t  
    
      

5. **Uniform Distribution:X~U(a,b)**  
Uniform Distribution is a probability distribution of Random Variable X, which represents an experiment whose outcomes are equally likely to occur when X takes a value in a given interval **(a,b)**.<br/>
 Example - Let R.V. $X$ represent rolling a dice experiment. Probability of any face landing is same, i.e, $\frac{1}{6}$. But, if $X$ takes 0 or 7 probability is zero, because its interval is from 1(a) to 6(b).
Uniform Distribution can be both discrete and continuous.In a discrete uniform distribution, outcomes are discrete. In a continuous uniform distribution, outcomes are continuous and infinite.  

    Let $X_d/X_c$ be a discrete/continuous Uniform R.V, whose equally likely outcomes takes an interval (a,b), then it's **pmf/pdf** can be defined as:  
$f(x) = \begin{cases} \frac{1}{b-a+1}& ,X_d=[a, a+1,...,b]\\0 & ,otherwise\end{cases}\hspace{0.5cm}$(for discrete) 
  
$\hspace{1.7cm}=\begin{cases} \frac{1}{b-a}& ,~X_c=[a,b]\\0 & ,otherwise\end{cases} \hspace{0.5cm}$(for continuous)  
  
* Expectation of discrete Uniform    
    $E[X_d] = \sum^b_{x=a} x \cdot ~ \frac{1}{b-a+1}$  
    $\hspace{1.3cm}= \frac{1}{b-a+1} \cdot \{a, a+1, a+2, ...., b\}$  
    $\hspace{1.3cm}= \frac{1}{b-a+1} \cdot \{a-a, a+1-a, a+2-a, ...., b-a\hspace{0.5cm} + \hspace{0.5cm} a(b-a+1)\}$  
    $\hspace{1.3cm}=\frac{1}{b-a+1} \cdot \{0, 1, 2, ...., b-a\hspace{0.5cm} + \hspace{0.5cm} a(b-a+1)\}$  
    $\hspace{1.3cm}= \frac{1}{b-a+1} \cdot \{0, 10, 2, ...., b-a\hspace{0.5cm} + \hspace{0.5cm} a(b-a+1)\}$  
    $\hspace{1.3cm}= \frac{1}{b-a+1} \cdot \{(b-a)(b-a+1)/2 + \hspace{0.5cm} a(b-a+1)\}$  
    $\hspace{1.3cm}= \frac{(b-a) + 2a}{2}$  
    $\hspace{1.3cm}= \frac{b+a}{2}$
      
      
* Expectation of continuous Uniform distribution  
    $E[X_c] = \int_a^b x \cdot \frac{1}{b-a}$  
    $\hspace{1.3cm}= \frac{b+a}{2}$  
    
      
* Variance of Discrete Uniform distribution  
    $Var[X_d] = \frac{(b-a+1)^2 ~-~ 1}{12}$  
    
      
* Variance of Continuous Uniform distribution  
    $Var[X_c] = \frac{(b-a)^2}{12}$  
    
      
* Moment Generating function discrete Uniform distribution  
    $\varphi_{X_d}(t) = \frac{e^{ta} - e^{t(b+1)}}{(b-a+1)(1- e^t)}$  
    
      
* Moment Generating function Continuous Uniform distribution  
    $\varphi_{X_c}(t) = \int_a^b \frac{e^{tx}}{b-a} dx~~~=\frac{e^{tb} - e^{ta}}{b-a}$
    

6. **Normal Distribution:X~N($\mu, \sigma^2$)**  
Normal Distribution, also known as the Gaussian distribution, is a probability distribution that is symmetric about the mean($\mu$), showing that data near the mean are more frequent in occurrence than data far from the mean, represented using a continuous Random Variable.<br/>
 Normal distributions are important in statistics and are often used in the natural and social sciences to represent real-valued random variables whose distributions are not known.  

    Let $X$ be a R.V of Normal Distribution, whose mean is $\mu$ and variance is $\sigma^2$, then it's **pdf** can be defined as:  
$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left( \frac{x-\mu}{\sigma}\right)^2}$  
  
* Expectation of Normal    
    $E[X] = \mu$
      
      
    
* Variance of Normal  
    $Var[X] = \sigma^2$  
    
      
* Moment Generating Function of Normal  
    $\varphi_X(t) = exp(\mu t ~+~ \sigma^2t^2/2)$
