## Sampling Distribution

A sampling distribution refers to a probability distribution of a statistic that comes from choosing random samples of a given population.The sampling distribution depends on multiple factors – 
* the statistic, 
* sample size, 
* sampling process, and 
* the overall population.
  
It is used to **estimate** statistics( like means,variances, and standard deviations) of the whole population using multiple samples.  
Since populations are typically large in size, it is important to use a sampling distribution so that you can randomly select a subset of the entire population.  
Sampling distribution also helps make the data easier to manage and builds a foundation for statistical inferencing, which leads to making inferences for the whole population.  
(As you might know, this is the main working principle of any ML model training).

**Steps to plot Sampling distribution -**  
1. Select a random sample($X$) of a specific size ($n$) from a given population.
2. Calculate a statistic for the sample, for example sample mean $\overline{X}$. (can be mode, median, or standard deviation as well)
3. Repeat the above two steps for numerous($m$) times keeping the sample size fixed, alongside keep track of the frequency of the calculated sample means$(\overline{X}_1, \overline{X}_2, \overline{X}_3, ......\overline{X}_m)$.
4. Plot probability distribution of the above frequency table, where x-axis represents the sample mean, and y-axis represent its probability of occurence($\frac{its~freq}{m(total~trials)}$ )
5. The plot you get is the sampling distribution.  

According to **Central Limit Theorem**, sampling distribution will turn out to be a Normal Distribution, with increase in no. of trials(or samplings).

Some facts when the 'statistic' of sample, used to generate sampling distribution, is mean $\overline{X}$:
* Mean of the sampling distribution will be almost same as the mean of the population distribution.
* Standard Error(same as sd) of sampling distribution will be - s.d. of population $/$ √ sample size taken in each trial.  

*Yes, we are calculating mean of sample means to estimate population mean*

**Central Limt Theorem**  
  
CLT states that when identically independent random variables are summed up, their properly normalized **Sum** tends toward a normal distribution even if the original variables themselves are not normally distributed.    
  
If $ X_{1},X_{2},\dots ,X_{m}$ are $m$ identically independent random variables of each sample(sample size =$n$) drawn $m$ times, from a population with overall mean $\mu$ and finite variance $\sigma ^{2}$,  
and if   
${\bar {X}}_{m} = \frac{\overline{X}_1 + \overline{X}_2 + \overline{X}_3 + ......+\overline{X}_m}{m}~$ is the Sampling Distribution of means(same as step 4),  
  
then the limiting form of the distribution :    
$Z=\lim _{m\to \infty }{\sqrt {n}}{\left({\frac {{\bar {X}}_{m}-\mu }{\sigma }}\right)}$, is a Standard Normal Distribution,  

or we can say ${\bar {X}}_{m}$ is a Normal Distribution with Mean $\mu$ and Standard Deviation $\frac{\sigma}{\sqrt {n}}$, when $m \rightarrow \infty$.
